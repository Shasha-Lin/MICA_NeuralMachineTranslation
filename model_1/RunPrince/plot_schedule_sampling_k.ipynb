```python
import matplotlib.pyplot as plt
import numpy as np
k = 3000
k/(k+np.exp(0/k))
x = np.arange(0, 50000, 5)
y = [k/(k+np.exp(i/k)) for i in x]
plt.plot(x, y)
plt.xlabel('epochs')
plt.ylabel('teacher forcing ratio according to schedule')
title = 'scheduled_sampling_k=%d'%k
plt.title(title)
plt.savefig(title+'.png')
plt.show()
```
